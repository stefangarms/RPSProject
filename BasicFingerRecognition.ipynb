```python
import cv2 # open cv library
import time # 
import os # 
import HandTrackingModule as htm
# import mediapipe as mp
```

```python
# define width and height of camera image
# wCam, hCam = 640, 480
wCam, hCam = 1080, 720

# turn on webcam and capture it
cap = cv2.VideoCapture(1)

# setting camera width and height
cap.set(3, wCam)
cap.set(4, hCam)

# store finger example images
ImagesPath = "Images"
if os.path.isfile(f'{ImagesPath}/.DS_Store'):
    os.remove(f'{ImagesPath}/.DS_Store')

myList = sorted(os.listdir(ImagesPath))
print(myList)
overlayList = [] #to overlay an image of the recognized fingers on the camera window
# create pathes for each image

for imPath in myList:
    image = cv2.imread(f'{ImagesPath}/{imPath}')
    # print(f'{ImagesPath}/{imPath}')
    
    # attach images to the List 'overlayList'
    overlayList.append(image)

# check if all images are imported
#print(len(overlayList))

# initialize previous Time
pTime = 0

# initialize detector
detector = htm.handDetector(detectionCon=0.75)

# initialize Finger Tip-ID's for Finger tip tracking
tipID = [4, 8, 12, 16, 20]
```

```python
while True:
    
    # Read the image
    success, img = cap.read()

    # display hand detection on image by using the created HandTrackingModule
    img = detector.findHands(img)

    # get Coordinates of Landmarks
    lmList = detector.findPosition(img, draw=False)
    fingers = []
    # check if List is not empty
    if len(lmList) != 0:
        #print(lmList)
        # if lmList[8][2] < lmList[6][2]:
        #     print("Zeigefinger")
        # else:
        #     print("geschlossen")

        

        if lmList[tipID[0]][1] < lmList[tipID[0]-1][1]:
            fingers.append(1)
        else:
            fingers.append(0)

        for id in range(1,5):
            if lmList[tipID[id]][2] < lmList[tipID[id]-2][2]:
                fingers.append(1)
            else:
                fingers.append(0)
        # print(fingers)

        # count open fingers by checking how many 1 in List fingers
        totalFingers = fingers.count(1)
        print(totalFingers)





    # flip image
    img = cv2.flip(img, 1)

    

    # place overlay picture in main image in top left corner (200... max values)
    h,w,c = overlayList[0].shape
    if len(lmList) != 0:
        img[0:h, 0:w] = overlayList[totalFingers-1]
        # show which fingers are shown on image
        cv2.putText(img, f'Open fingers: {int(fingers[0])} {int(fingers[1])} {int(fingers[2])} {int(fingers[3])} {int(fingers[4])}', (500,650), cv2.FONT_HERSHEY_PLAIN, 2, (255,255,0), 2)

    # display frame rate
    cTime = time.time()
    fps = 1/(cTime-pTime)
    pTime = cTime
    cv2.putText(img, f'FPS: {int(fps)}', (400,70), cv2.FONT_HERSHEY_PLAIN, 3, (255,0,0),3)


    # Display the resulting image
    cv2.imshow("Image", img)
    # cv2.waitKey(1)
    if cv2.waitKey(5) & 0xFF == 27:
        cv2.destroyAllWindows()
        print("quit")
        break
```
